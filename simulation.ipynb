```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import viz # curriculum example visualizations

np.random.seed(29)
```

```python
## set the number of trials 

n_trials = 10_000
n_rows= 10_000

##roll three dice

n_dice = 3 
n_cols = 3

#use np's random choice 
# select from [1,2,3,4,5,6]

rolls = np.random.choice([1,2,3,4,5,6], n_trials * n_dice).reshape(n_rows, n_cols)

rolls
```

```python
#Find the sum of each dice roll
# Sum with axis = 1
sum_by_trial = rolls.sum(axis = 1)
```

```python
def simulation_example1(sums_by_trial):
    freq, bins = np.histogram(sums_by_trial, bins=range(2, 21))
    freq = freq / freq.sum() # raw number -> percent
    plt.bar(bins[:-1], freq, width=1, color='white', edgecolor='black')
    plt.bar(bins[-8:-1], freq[-7:], width=1, color='lightblue', edgecolor='black')

    plt.xticks(range(3, 19))
    plt.ylabel('% of the time we get this sum')
    plt.xlabel('Sum of rolls')

    num_rolls = '{:,}'.format(len(sums_by_trial))
    plt.title(f'Sum of 3 dice rolls based on {num_rolls} simulations')
    plt.xlim(1, 23)
    plt.annotate('', xy=(13, .11), xytext=(18, .11), xycoords='data',
                 textcoords='data', arrowprops={'arrowstyle': '|-|'})
    plt.text(14.5, .12, 'P(X > 12)', ha='left', va='center')
```

```python
simulation_example1(sum_by_trial)
```

```python
#check the condition > 12
wins = sum_by_trial > 12
```

```python
wins
```

```python
#set win array to integers (0 or 1)
# find the mean to find the proportion of wins
win_rate = wins.astype(int).mean()
```

```python
win_rate
print(f'The player is expected to win {win_rate * 100}% of the time.')
```

```python
#calculate the expected winnings
expected_winnings = win_rate * 15

#account for the price of the game
cost = 5

#profit
expected_profit = expected_winnings - cost
```

```python
expected_profit
```

Codeup students buy, on average, 3 poptart packages with a standard deviation of 1.5 a day from the snack vending machine. If on monday the machine is restocked with 17 poptart packages, how likely is it that I will be able to buy some poptarts on Friday afternoon? (Remember, if you have mean and standard deviation, use the np.random.normal) You'll need to make a judgement call on how to handle some of your values

```python
### Normal distribution of poptarts purchases
```

```python
mean = 3 
sd = 1.5

#sales over 5 days 
n_days = 5
n_trials = 10000
```

```python
poptart_sales = np.random.normal(mean,sd,size=(n_trials, n_days))
```

```python
poptart_sales
```

```python
poptart_sales[poptart_sales < 0] = 0
```

```python
poptart_weekly_sales = poptart_sales.sum(axis = 1)
```

```python
poptart_weekly_sales
```

```python
(poptart_weekly_sales < 17).mean()
```

When installing anaconda on a student's computer, there's a 1 in 250 chance that the download is corrupted and the installation fails. What are the odds that after having 50 students download anaconda, no one has an installation issue? 100 students?

What is the probability that we observe an installation issue within the first 150 students that download anaconda?

```python
outcomes = ["S", 'F']

n_students = 50
```

```python
downloads = np.random.choice(outcomes, size = (10000, n_students), p = [249/250, 1/250])
```

```python
downloads
```

```python
((downloads == 'S').sum(axis = 1)==50).mean()
```

```python
#Same set up just checking for 100 students instead of 50 
n_students = 100
downloads = np.random.choice(outcomes, size = (10000, n_students), p = [249/250, 1/250])
```

```python
((downloads == 'S').sum(axis = 1)==100).mean()
```

```python
#checking for at least 1 failure
## 150 installations

downloads = np.random.choice(outcomes, size = (10000, 150), p = [249/250, 1/250])
```

```python
bools = downloads == 'S'
```

```python
number_successful_installations = bools.sum(axis=1)
```

```python
number_successful_installations != 150
```

```python
(number_successful_installations != 150).mean()
```
